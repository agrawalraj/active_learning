```python
import numpy as np
import sampler
import graph_utils
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Generate data 
p = 5
edge_prob = .9
g = graph_utils.random_graph(p, edge_prob)
adj_mat = graph_utils.random_adj(g)
prec_mat = graph_utils.adj2prec(adj_mat)
int_data = graph_utils.sample_graph_int(g, adj_mat, [0, 1, 2, 3], [50, 50, 50, 50])
```

```python
_, cheat_run1 = sampler.sample_dags(g, prec_mat, int_data, 100, 1, 1000)
```

```python
_, cheat_run2 = sampler.sample_dags(g, prec_mat, int_data, 1000, 1, 10000)
```

```python
ax1 = plt.subplot(131)
ax1.plot(cheat_run1)
ax1.set_ylim([-8585, -8570])
```

```python
ax1 = plt.subplot(131)
ax1.plot(cheat_run2)
ax1.set_ylim([-8585, -8570])
```
